```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import pybaseball
import numpy as np
import pandas as pd
from scipy import stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# regular season

```python
url = 'https://baseballsavant.mlb.com/leaderboard/pitch-tempo?game_type=Regular&n=q&season_end=2024&season_start=2020&split=yes&team=&type=Pit&with_team_only=1'

# Send a GET request to fetch the webpage content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all tables on the page
tables = soup.find('div', class_='article-template')


# Check if any tables are found
tmp = tables.find('script').string
```

```python
script_content = tables.find('script').string

# Split the content into lines and strip each line
lines = [line.strip() for line in script_content.split('\n')]
```

```python
df = pd.DataFrame(json.loads(lines[2:][1][6:-1]))
```

```python
drop_col = ['start_year_lower','start_year','end_year','end_year_upper','key_name_display_club','num_teams','key_team_id','pitcher_searchable','key_file_code','key_name_abbrev','team_image_url']
```

```python
df2 = df.loc[:, df.nunique() > 1]
df3 = df2.drop(columns=drop_col)
df3.rename(columns={'timeframe':'year'},inplace=True)
reg_df = df3.sort_values(by='entity_name')
reg_df['game_type'] = 'Regular'
reg_df.head()
```

```python
reg_df.groupby('year')['tot_n'].min()
```

# Play off

```python
url = 'https://baseballsavant.mlb.com/leaderboard/pitch-tempo?game_type=Playoff&n=q&season_end=2024&season_start=2020&split=yes&team=&type=Pit&with_team_only=1'
# Send a GET request to fetch the webpage content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all tables on the page
tables = soup.find('div', class_='article-template')

# Check if any tables are found
tmp = tables.find('script').string
script_content = tables.find('script').string
# Split the content into lines and strip each line
lines = [line.strip() for line in script_content.split('\n')]
df = pd.DataFrame(json.loads(lines[2:][1][6:-1]))
```

```python
df2 = df.loc[:, df.nunique() > 1]
df3 = df2.drop(columns=drop_col)
df3.rename(columns={'timeframe':'year'},inplace=True)
po_df = df3.sort_values(by='entity_name')
po_df['game_type'] = 'Playoff'
po_df.head()
```

# merge

```python
timer_df = pd.concat([reg_df,po_df])
```
