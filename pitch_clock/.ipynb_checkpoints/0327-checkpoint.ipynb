# Data load

```python
import pandas as pd
import pybaseball
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 기본 성적지표
classic_data = pybaseball.pitching_stats(start_season= 2021, end_season = 2024,qual=10) # 최소이닝수 10
classic_data_v2 = pybaseball.pitching_stats(start_season= 2021, end_season = 2024)
```

```python
# timer 데이터
timer_data = pd.read_csv('data/timer_data.csv')
timer_data = timer_data[timer_data['year']>2020]
timer_rg = timer_data[timer_data['game_type']=='Regular']
```

```python
pitcher_id_v1 = timer_rg.groupby(['entity_id'])['year'].count()[timer_rg.groupby(['entity_id'])['year'].count() ==4].index
```

```python
classic_v1 = classic_data[classic_data['IDfg'].isin(pybaseball.playerid_reverse_lookup(player_ids=list(pitcher_id_v1))['key_fangraphs'])]
timer_v1 = timer_rg[timer_rg['entity_id'].isin(pitcher_id_v1)]
timer_v1 = timer_v1.rename(columns={'year':'Season'})
```

```python
ID_dict = {} # Fg -> mlb
for i in pitcher_id_v1:
    mlb_id = i
    fg_id = pybaseball.playerid_reverse_lookup([mlb_id])['key_fangraphs'][0]
    ID_dict[int(fg_id)] = mlb_id
```

```python
classic_v1['entity_id'] = classic_v1['IDfg'].map(ID_dict)
```

```python
merg_v1 = pd.merge(classic_v1,timer_v1,on=['entity_id','Season'],how='inner')
```

# Grouping

```python
for p_id in pitcher_id_v1:
    if merg_v1.loc[(merg_v1['entity_id']==p_id)&(merg_v1['Season']==2021),'median_seconds'].values[0] <18.81:
        merg_v1.loc[merg_v1['entity_id']==p_id,'g_21'] = 'Fast'
        
    elif merg_v1.loc[(merg_v1['entity_id']==p_id)&(merg_v1['Season']==2021),'median_seconds'].values[0] <20.57:
        merg_v1.loc[merg_v1['entity_id']==p_id,'g_21'] = 'Middle'
         
    else:
        merg_v1.loc[merg_v1['entity_id']==p_id,'g_21'] = 'Slow'

    if merg_v1.loc[(merg_v1['entity_id']==p_id)&(merg_v1['Season']==2022),'median_seconds'].values[0] <18.60:
        merg_v1.loc[merg_v1['entity_id']==p_id,'g_22'] = 'Fast'
        
    elif merg_v1.loc[(merg_v1['entity_id']==p_id)&(merg_v1['Season']==2022),'median_seconds'].values[0] <20.82:
        merg_v1.loc[merg_v1['entity_id']==p_id,'g_22'] = 'Middle'
         
    else:
        merg_v1.loc[merg_v1['entity_id']==p_id,'g_22'] = 'Slow'

    if merg_v1.loc[(merg_v1['entity_id']==p_id)&(merg_v1['Season']==2023),'median_seconds'].values[0] <15.83:
        merg_v1.loc[merg_v1['entity_id']==p_id,'g_23'] = 'Fast'
        
    elif merg_v1.loc[(merg_v1['entity_id']==p_id)&(merg_v1['Season']==2023),'median_seconds'].values[0] <17.11:
        merg_v1.loc[merg_v1['entity_id']==p_id,'g_23'] = 'Middle'
         
    else:
        merg_v1.loc[merg_v1['entity_id']==p_id,'g_23'] = 'Slow'
```

## t-test

```python
# 2021년 Slow 그룹과 Fast 그룹 t-test
```

```python
# 정규성 검증
sns.histplot(data=merg_v1[merg_v1['Season']==2021],x='ERA',hue='g_21',kde=True,palette='tab10')
plt.show()
```

```python
merg_v1['Season']==2021
```

```python
merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Slow"),'ERA'].values
```

### 2021

```python
# Slow vs Fast
from scipy import stats

group1 = merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Slow"),'ERA'].values
group2 = merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Fast"),'ERA'].values

t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_value:.3f}")
```

```python
# Slow vs Middle
from scipy import stats

group1 = merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Slow"),'ERA'].values
group2 = merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Middle"),'ERA'].values

t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_value:.3f}")
```

```python
# Fast vs Middle
from scipy import stats

group1 = merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Fast"),'ERA'].values
group2 = merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Middle"),'ERA'].values

t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_value:.3f}")
```

```python
group1 = merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Fast"),'ERA'].values
group2 = merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Middle"),'ERA'].values
group3 = merg_v1.loc[(merg_v1['Season']==2021)&(merg_v1['g_21']=="Slow"),'ERA'].values

f_stat, p_value = stats.f_oneway(group1, group2, group3)

print(f"F-statistic: {f_stat:.3f}")
print(f"p-value: {p_value:.3f}")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd

# 2021년 데이터에서 필요한 열 추출
df_2021 = merg_v1.loc[merg_v1['Season'] == 2021, ['g_21', 'ERA']].dropna()

# Tukey HSD 수행
tukey = pairwise_tukeyhsd(endog=df_2021['ERA'], groups=df_2021['g_21'], alpha=0.05)

print(tukey)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd

# 2021년 데이터 필터링
df_2021 = merg_v1.loc[merg_v1['Season'] == 2021, ['g_21', 'ERA']].dropna()

# ANOVA 모델 적합
model = ols('ERA ~ C(g_21)', data=df_2021).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# 결과 출력
print("\n[ANOVA Table]")
print(anova_table)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Tukey 사후분석
tukey = pairwise_tukeyhsd(endog=df_2021['ERA'], groups=df_2021['g_21'], alpha=0.05)

# 결과 DataFrame으로 변환
tukey_df = pd.DataFrame(data=tukey.summary().data[1:], columns=tukey.summary().data[0])

print("\n[Tukey HSD Table]")
print(tukey_df)
```
