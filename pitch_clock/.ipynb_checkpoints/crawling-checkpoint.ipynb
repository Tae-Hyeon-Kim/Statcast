```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import pybaseball
import numpy as np
import pandas as pd
from scipy import stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Crawling

## regular season

```python
url = 'https://baseballsavant.mlb.com/leaderboard/pitch-tempo?game_type=Regular&n=q&season_end=2024&season_start=2020&split=yes&team=&type=Pit&with_team_only=1'

# Send a GET request to fetch the webpage content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all tables on the page
tables = soup.find('div', class_='article-template')


# Check if any tables are found
tmp = tables.find('script').string
```

```python
script_content = tables.find('script').string

# Split the content into lines and strip each line
lines = [line.strip() for line in script_content.split('\n')]
```

```python
df = pd.DataFrame(json.loads(lines[2:][1][6:-1]))
```

```python
drop_col = ['start_year_lower','start_year','end_year','end_year_upper','key_name_display_club','num_teams','key_team_id','pitcher_searchable','key_file_code','key_name_abbrev','team_image_url']
```

```python
df2 = df.loc[:, df.nunique() > 1]
df3 = df2.drop(columns=drop_col)
df3.rename(columns={'timeframe':'year'},inplace=True)
reg_df = df3.sort_values(by='entity_name')
reg_df.head()
```

```python
rg_tp_df = reg_df.groupby(['entity_name','entity_id','year'])['median_seconds_empty'].mean().reset_index()
rg_tp_df = rg_tp_df.pivot(index=['entity_name','entity_id'],columns='year',values='median_seconds_empty').reset_index()#.sort_values(by='entity_name')
rg_tp_df.columns.name = ''
#rg_tp_df = rg_tp_df[~rg_tp_df[2023].isna()]
rg_tp_df.head()
```

```python
rg_tp_df.columns = ['entity_name', 'entity_id', '2020', '2021', '2022', '2023', '2024']
```

## po

```python
url = 'https://baseballsavant.mlb.com/leaderboard/pitch-tempo?game_type=Playoff&n=q&season_end=2023&season_start=2020&split=yes&team=&type=Pit&with_team_only=1'
# Send a GET request to fetch the webpage content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all tables on the page
tables = soup.find('div', class_='article-template')

# Check if any tables are found
tmp = tables.find('script').string
script_content = tables.find('script').string
# Split the content into lines and strip each line
lines = [line.strip() for line in script_content.split('\n')]
df = pd.DataFrame(json.loads(lines[2:][1][6:-1]))
```

```python
df2 = df.loc[:, df.nunique() > 1]
df3 = df2.drop(columns=drop_col)
df3.rename(columns={'timeframe':'year'},inplace=True)
po_df = df3.sort_values(by='entity_name')
po_df.head()
```

```python
po_tp_df = po_df.groupby(['entity_id','entity_name','year'])['median_seconds_empty'].mean().reset_index()
po_tp_df = po_tp_df.pivot(index=['entity_name','entity_id'],columns='year',values='median_seconds_empty').reset_index()#.sort_values(by='entity_name')
po_tp_df.columns.name = ''
po_tp_df
```

## get pitching data

```python
id_list = list(rg_tp_df['entity_id'].values)
```

```python
stats_20 = pybaseball.pitching_stats_bref('2020')
stats_21 = pybaseball.pitching_stats_bref('2021')
stats_22 = pybaseball.pitching_stats_bref('2022')
stats_23 = pybaseball.pitching_stats_bref('2023')
```

```python
stats_20
```

```python
stats_20['mlbID'] = stats_20['mlbID'].astype(int)
stats_21['mlbID'] = stats_21['mlbID'].astype(int)
stats_22['mlbID'] = stats_22['mlbID'].astype(int)
stats_23['mlbID'] = stats_23['mlbID'].astype(int)
```

```python
stats_20 = stats_20[stats_20['mlbID'].isin(id_list)].sort_values(by='Name').reset_index(drop=True)
stats_21 = stats_21[stats_21['mlbID'].isin(id_list)].sort_values(by='Name').reset_index(drop=True)
stats_22 = stats_22[stats_22['mlbID'].isin(id_list)].sort_values(by='Name').reset_index(drop=True)
stats_23 = stats_23[stats_23['mlbID'].isin(id_list)].sort_values(by='Name').reset_index(drop=True)

stats_21.head()
```

# Analysis

## 이전 시즌과 비교하여 피치템포가 정말 줄었는지?

### 2022 시즌과 2023 시즌 비교

```python
tmp_df = rg_tp_df[(rg_tp_df['2023']==rg_tp_df['2023'])&(rg_tp_df['2022']==rg_tp_df['2022'])]
tmp_df
```

```python
t_statistic, p_value = stats.ttest_rel(tmp_df['2022'], tmp_df['2023'])

print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
```

### 2020~2022 시즌과 2023 시즌 비교

```python
tmp_df = rg_tp_df.copy()
tmp_df['avg'] = rg_tp_df[['2020','2021','2022']].mean(axis=1)
tmp_df = tmp_df[(~tmp_df['avg'].isna())&(~tmp_df['2023'].isna())]
tmp_df
```

```python
t_statistic, p_value = stats.ttest_rel(tmp_df['avg'], tmp_df['2023'])

print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
```

## 피치템포와 성적간의 상관관계

```python
corr_columns = [ 'Age', '#days' ,'G', 'GS', 'W', 'L', 'SV', 'IP',
       'H', 'R', 'ER', 'BB', 'SO', 'HR', 'HBP', 'ERA', 'AB'
       ,'GDP', 'SF', 'SB', 'CS', 'PO', 'BF', 'Pit', 'Str', 'StL', 'StS',
       'GB/FB', 'LD', 'PU', 'WHIP', 'BAbip', 'SO9', 'SO/W']
```

```python
stats_21.columns
```

```python
merge_20 = pd.merge(stats_20.rename(columns={'mlbID':'entity_id'}),rg_tp_df[['entity_id','2020']],on='entity_id')
merge_20 = merge_20[~merge_20['2020'].isna()]
s20 = merge_20[corr_columns+['2020']].corr()['2020']
```

```python
merge_21 = pd.merge(stats_21.rename(columns={'mlbID':'entity_id'}),rg_tp_df[['entity_id','2021']],on='entity_id')
merge_21 = merge_21[~merge_21['2021'].isna()]
s21 = merge_21[corr_columns+['2021']].corr()['2021']
```

```python
merge_22 = pd.merge(stats_22.rename(columns={'mlbID':'entity_id'}),rg_tp_df[['entity_id','2022']],on='entity_id')
merge_22 = merge_22[~merge_22['2022'].isna()]
s22 = merge_22[corr_columns+['2022']].corr()['2022']
```

```python
merge_23 = pd.merge(stats_23.rename(columns={'mlbID':'entity_id'}),rg_tp_df[['entity_id','2023']],on='entity_id')
merge_23 = merge_23[~merge_23['2023'].isna()]
s23 = merge_23[corr_columns+['2023']].corr()['2023']
```

```python
df = pd.concat([s20, s21, s22, s23], axis=1)

# Generate the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, cmap='coolwarm', linewidths=.5,vmax=1,vmin=-1)
#plt.title('Heatmap of Combined Series')
plt.show()
```

```python
sns.regplot(data=merge_21,x='2021',y='G',line_kws=dict(color="r"))
plt.xlabel('2021 Pitch Tempo')
plt.show()
```

```python
sns.regplot(data=merge_23,x='2023',y='G',line_kws=dict(color="r"))
plt.xlabel('2023 Pitch Tempo')
plt.show()
```

## 피치템포가 줄어든 폭과 성적간의 상관관계

### 직전시즌(2022)과 비교

```python
tmp_df = rg_tp_df[(rg_tp_df['2023']==rg_tp_df['2023'])&(rg_tp_df['2022']==rg_tp_df['2022'])]
tmp_df['diff'] = tmp_df['2022'] - tmp_df['2023']
diff_df = pd.merge(stats_23.rename(columns={'mlbID':'entity_id'}),tmp_df[['entity_id','diff']],on='entity_id')
diff_sr = diff_df[corr_columns+['diff']].corr()['diff']
```

```python
diff_sr
```

### 20-22 시즌 템포 평균과 2023시즌 템포와 비교

```python
tmp_df = rg_tp_df[(rg_tp_df['2023']==rg_tp_df['2023'])&(rg_tp_df['2022']==rg_tp_df['2022'])]
tmp_df['diff'] = rg_tp_df[['2020','2021','2022']].mean(axis=1) - tmp_df['2023']
diff_df = pd.merge(stats_23.rename(columns={'mlbID':'entity_id'}),tmp_df[['entity_id','diff']],on='entity_id')
diff_sr = diff_df[corr_columns+['diff']].corr()['diff']
```

```python
tmp_col = abs(diff_sr).sort_values().index[-10:]
```

```python
diff_df
```

```python
plt.figure(figsize=(10,10))
sns.pairplot(data=diff_df[tmp_col])
plt.show()
```

## 2023년 성적이 떨어진 선수들의 이전 피치템포

```python
inter_22_23 = list(set(stats_23.mlbID).intersection(set(stats_22.mlbID)))
```

```python
id_23 = rg_tp_df.loc[~rg_tp_df['2023'].isna(),'entity_id']
set23 = set(stats_22.loc[stats_22['mlbID'].isin(id_23),'mlbID'])
set22 = set(stats_23.loc[stats_23['mlbID'].isin(id_23),'mlbID'])
inter_22_23 = list(set23.intersection(set22))
```

```python
stats22_v2 = stats_22[stats_22['mlbID'].isin(inter_22_23)].reset_index(drop=True)
stats23_v2 = stats_23[stats_23['mlbID'].isin(inter_22_23)].reset_index(drop=True)
```

```python
stats23_v2['ERA_diff'] = stats22_v2['ERA'] - stats23_v2['ERA']
```

```python
tmp_df = stats23_v2.sort_values(by='ERA_diff')[['mlbID','ERA_diff']]
tmp_df
```

```python
tmp_df.rename(columns={'mlbID':'entity_id'},inplace=True)
era_diff_df = pd.merge(rg_tp_df,tmp_df,on='entity_id',how='inner').sort_values(by='ERA_diff')
era_diff_df
```

```python
# 2022년 피치템포와 2022년과 2023년 ERA차이의 상관관계

sns.regplot(data = era_diff_df[~era_diff_df['2022'].isna()],x='2022',y='ERA_diff')
plt.show()
```

```python
era_diff_df[['2020','2021','2022']].mean(axis=1)
```

```python
# 2022년 이전 피치템포 평균과 2022년과 2023년 ERA 차이의 상관관계

era_diff_df['20_22_mean'] = era_diff_df[['2020','2021','2022']].mean(axis=1)
sns.regplot(data = era_diff_df[~era_diff_df['20_22_mean'].isna()],x='20_22_mean',y='ERA_diff')
plt.show()
```
