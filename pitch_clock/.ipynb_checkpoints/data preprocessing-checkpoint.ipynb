```python
import pandas as pd
import numpy as np
import pybaseball
import pickle
```

```python
rg_24 = pd.read_csv('data/pitch_tempo_24.csv')
rg_23 = pd.read_csv('data/pitch_tempo_23.csv')
rg_22 = pd.read_csv('data/pitch_tempo_22.csv')
rg_21 = pd.read_csv('data/pitch_tempo_21.csv')
rg_20 = pd.read_csv('data/pitch_tempo_20.csv')

po_23 = pd.read_csv('data/pitch_tempo_23_po.csv')
po_22 = pd.read_csv('data/pitch_tempo_22_po.csv')
po_21 = pd.read_csv('data/pitch_tempo_21_po.csv')
po_20 = pd.read_csv('data/pitch_tempo_20_po.csv')
```

```python
rg_24['year'] = 2024
rg_23['year'] = 2023
rg_22['year'] = 2022
rg_21['year'] = 2021
rg_20['year'] = 2020

po_23['year'] = 2023
po_22['year'] = 2022
po_21['year'] = 2021
po_20['year'] = 2020
```

```python
rg_data = pd.concat([rg_24,rg_23,rg_22,rg_21,rg_20])
rg_data.head()
```

```python
over_23_name = list(rg_data[rg_data['year']>=2023]['entity_name'].unique())
under_23_name = list(rg_data[rg_data['year']<2023]['entity_name'].unique())
```

```python
name_list = list(set(under_23_name)&set(over_23_name))
```

```python
# print(rg_data['entity_name'].nunique())
# print(len(name_list))
rg_df = rg_data[rg_data['entity_name'].isin(name_list)]
rg_df['entity_name'].nunique()
```

```python
rg_data.groupby('year')['median_seconds_empty'].mean().reset_index()
```

```python
rg_tp_df = rg_df.groupby(['entity_name','year'])['median_seconds_empty'].mean().reset_index()
rg_tp_df = rg_tp_df.pivot(index='entity_name',columns='year',values='median_seconds_empty').reset_index()
rg_tp_df.columns.name = ''
rg_tp_df
```
